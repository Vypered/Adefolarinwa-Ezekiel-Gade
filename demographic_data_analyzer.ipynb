```python
#importing python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
#Importing the dataset: adult.data(32,561 entries)
data = pd.read_csv(r'C:\Users\USERPC\PyCharmMiscProject\adult\adult.data',header=None)
#Renaming the columns
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary']
data.columns = column_names
total_figure  = 32561
data.tail(20)
```

```python
#Detailed information about our data
data.info()
```

```python
#To check for null values
data.isnull().sum()
#result indicating there is no null value among all columns
```

### How many people of each race are represented?


```python
#The race column is to be queried to determine the value counts of each race.
plt.figure(figsize=(20,6))
plt.grid('whitegrid')
sns.histplot(x= data['race'], data= data, label= 'race', hue=  data['race'], palette= 'Set1')
for p in plt.gca().patches:
    height = p.get_height()
    plt.text(p.get_x() + p.get_width()/2, height,str(int(height)), ha='center', va='bottom')
```

### Average age of men?


```python
#To determine the average age of men.
average_age = round(data['age'].mean())
print(f'The average men\'s age is {average_age} years')
```

### Percentage of people with Bachelor's degree?


```python
#To determine the percentage of people with Bachelor's degree, we have to determine the overall qualification value counts from the 32,561 entries.
education_percent = data['education'].value_counts()
plt.figure(figsize=(20,6))
plt.pie(education_percent, autopct= '%.1f%%', pctdistance= 1.2, colors= ('beige','silver','gold','coral','sienna','cyan','orange','skyblue','violet','red','purple', 'blue', 'forestgreen'),labels= ('HS-Grad','Some-College','Bachelors','Masters','Assoc-Voc','11th','Assoc-acdm','10th','7th-8th','Prof-School','9th','12th','Doctorate','5th-6th','1st-4th','PreSchool'), labeldistance= 1.3, rotatelabels= True, shadow= True)
plt.title('Advanced Learning[B(s), M(s), Dr] Earners')
```

```python
#From the above, men with Bachelor's degree = 5355
bachelors_count =  5355
bachelors_percent = round((bachelors_count/total_figure) * 100)
print(f'Percentage of people with Bachelor\'s degree: {bachelors_percent}%')
```

### Percentage of people with advanced education(Bachelors, Masters, Doctorate) that make more than $50,000?


```python
#All advanced learners including Bachelors, Masters and Doctorate.
all_advance_earners = data[(data['education-num'] > 12) & (data['salary'] == (' >50K'))]
print(f'There are {len(all_advance_earners)} advanced learners earning above $50,000.')
```

Calculations show that 3,909 that have advanced education(Bachelors, Masters, Prof-school, Doctorate) out of 34,561 people  earn above $50,000.

Indicating **12.005%** of the overall population.

```python
#Gender effect on salary
male_50k = data[(data['sex'] == ' Male') & (data['salary'] == (' >50K'))]
male_50k = len(male_50k)
female_50k = data[(data['sex'] == ' Female') & (data['salary'] == (' >50K'))]
female_50k = len(female_50k)
male_b50k = data[(data['sex'] == ' Male') & (data['salary'] == (' <=50K'))]
male_b50k = len(male_b50k)
female_b50k = data[(data['sex'] == ' Female') & (data['salary'] == (' <=50K'))]
female_b50k = len(female_b50k)
#Creating a table for the results
gender_salary =  {'Figures' : [male_50k, female_50k, male_b50k, female_b50k],
          'Details' : ['Male > $50,000','Female > $50,000','Male < $50,000','Female < $50,000']
        }
gender_salary
```

```python
plt.figure(figsize=(20,6))
plt.grid('whitegrid')
sns.barplot(x= gender_salary['Details'], y= gender_salary['Figures'], data= gender_salary, color= 'cyan', width= 0.5, hue= gender_salary['Details'], palette= 'Set1')
plt.title('Gender-Salary Distribution')
plt.xlabel('Gender and Earning')
plt.ylabel('Number of People')
for p in plt.gca().patches:
    height = p.get_height()
    plt.text(p.get_x() + p.get_width()/2, height,str(int(height)), ha='center', va='bottom')
```
