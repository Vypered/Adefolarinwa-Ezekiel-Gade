```python
import numpy as np
```

```python
def calculate(f1, f2, f3, f4, f5, f6, f7, f8, f9):
    num_list = np.array([[f1, f2, f3], [f4, f5, f6], [f7, f8, f9]])
    #Axis = 1(each columns)
    list_col_mean = num_list.mean(axis=0)
    list_col_std = num_list.std(axis=0)
    list_col_var = num_list.var(axis=0)
    list_col_sum = num_list.sum(axis=0)
    list_col_max = num_list.max(axis=0)
    list_col_min = num_list.min(axis=0)

    #Axis = 0( each rows)
    list_row_mean = num_list.mean(axis=1)
    list_row_std = num_list.std(axis=1)
    list_row_var = num_list.var(axis=1)
    list_row_sum = num_list.sum(axis=1)
    list_row_max = num_list.max(axis=1)
    list_row_min = num_list.min(axis=1)

    #Flattened
    list_mean = num_list.mean()
    list_std = num_list.std()
    list_var = num_list.var()
    list_sum = num_list.sum()
    list_max = num_list.max()
    list_min = num_list.min()

    print(f"""    'mean': [{list_col_mean}], [{list_row_mean}],[ {list_mean}]]
    'variance': [{list_col_var}], [{list_row_var}],[ {list_var}]],
    'standard deviation': [{list_col_std}], [{list_row_std}],[ {list_std}]],
    'max': [{list_col_max}], [{list_row_max}],[ {list_max}]],
    'min': [{list_col_min}], [{list_row_min}],[ {list_min}]],
    'sum': [{list_col_sum}], [{list_row_sum}],[ {list_sum}]]
    """)

```

```python
calculate(10,20,30,49,50,69,70,80,90)
```
